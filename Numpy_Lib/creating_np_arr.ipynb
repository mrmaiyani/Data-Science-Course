## 3. Changing Data Type

```python
krishna = np.array([1,2,3])
```

```python
krishna.dtype
```

```python
krishna1 = np.array([1,2,3], dtype ='float32')
# or krishna1 = np.array([1,2,3], dtype = np.float32)
krishna1.dtype
```

```python
# જ્યારે વેરિએબલ પહેલેથી જ બનાવેલ હોય અને તેના ડેટાટાઈપ માં ફેરફાર કરવો હોય ત્યારે:
krishna1_int = krishna1.astype(np.int32)
print(krishna1_int)
```
