```python
import numpy as np
import time
```

```python
size = 1_000_000_00

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(l1, l2)]
end = time.time()

print(end - start)
```

```python
size = 1_000_000_00

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()

print(end - start)
```

```python
list1 = [1,2,3]
start = time.time()
list_squares = [x**2 for x in list1]
end = time.time()
print(list_squares)
print(end - start)
```

```python
arr2 = np.array([1,2,3])
start = time.time()
numpy_squares = arr2 ** 2 
end = time.time()
print(numpy_squares)
print(end - start)
```
