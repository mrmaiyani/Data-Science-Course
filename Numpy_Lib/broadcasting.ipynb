```python
# normalization
data = np.array([[10, 20, 30],
                [15,25,35],
                [20, 30 ,40],
                [25,35 , 40],
                [30, 40, 50]])

mean = data.mean(axis = 0)
std = data.std(axis = 0)

normalized = (data - mean) / std

normalized
```
