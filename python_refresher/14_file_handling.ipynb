# File Handling

### Using Write() -Overwrite existing content

```python
file = open("15_file_handling","w")
file.write("Hello World")
file.close()
```

### Using WriteLines() -Write Multiple Lines

```python
lines = ["Hello\n", "Welcome to python\n","File handling\n"]

file = open("15_file_handling","w")
file.writelines(lines)
file.close()
```

### Using read() -Read entire file

```python
file = open("15_file_handling","r")
content = file.read()
print(content)
file.close()
```

### Using readline() -read line by line

```python
file = open("15_file_handling","r")
line1 = file.readline()
print(line1)
file.close()
```

### Using readlines() -read all lines as list

```python
file = open("15_file_handling","r")
lines = file.readlines()
print(lines)
file.close()
```

### appending to a file

```python
file = open("15_file_handling","a")
file.write("\nThis is an additional line.")
file.close()
```

### Using with Statement (Best Practice)

```python
 with open("15_file_handling", "r") as file:
    content = file.read()
 print(content) # No need to manually close the fil
```

###  Checking if a File Exists

```python
 import os
 if os.path.exists("15_file_handling"):
     print("File exists!")
else:
    print("File not found!")
```

### Deleting a File

```python
import os
if os.path.exists("example.txt"):
    os.remove("example.txt")
    print("File deleted.")
else:
    print("File does not exist.")
```
